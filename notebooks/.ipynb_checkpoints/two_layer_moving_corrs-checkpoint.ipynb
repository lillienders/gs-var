```python
fig_path  = '/Users/lillienders/Documents/gs_var/figures/'
data_path = '/Users/lillienders/Documents/gs_var/data/'
```

```python
time_orca = np.arange(np.datetime64(str(1958) + '-01-01'), np.datetime64(str(2016 + 1) + '-01-01'),
                 np.timedelta64(1, 'M'), dtype='datetime64[M]')
```

```python
nao_uf  = xr.open_dataarray(data_path + 'cli_indices/nao_era.nc')
nao_f = lanczos_bandpass(nao_uf, low_freq = 12*6, high_freq = 12*11, window = 12) 
nao_f = (nao_f - nao_f.mean(dim='time'))/nao_f.std(dim = 'time')
```

```python
gsi_uf_tl   = xr.open_dataarray(data_path + 'gsi/two_layer_gsi.nc')
gsi_f_tl    = lanczos_bandpass(gsi_uf_tl, low_freq = 12*6, high_freq = 12*11, window = 12) 
gsi_f_tl    = (gsi_f_tl - gsi_f_tl.mean(dim='time'))/gsi_f_tl.std(dim = 'time')
```

```python
gsi_f_tl_drop_na = gsi_f_tl[5:-5]
nao_f_drop_na = nao_f[5:-5]
```

```python
nao_f = nao_f_drop_na.sel(time = slice(nao_f_drop_na.time[0],gsi_f_tl_drop_na.time[-1]))
gsi_f = gsi_f_tl_drop_na.sel(time = slice(nao_f_drop_na.time[0],gsi_f_tl_drop_na.time[-1]))
```

```python
n_lags = 12*10
lagged_corrs_f = lagged_corr(nao_f, gsi_f, n_lags) # Calculate Lagged Correlations
lagged_corrs_f = np.array(lagged_corrs_f)          # Convert to array
lci_f,uci_f    = get_cis(nao_f, gsi_f,n_lags)      # Calculate confidence intervals
```

```python
# For plotting: Grab indices of points that exceed confidence threshold
ix_f = []
for l in range(len(lagged_corrs_f)):
        if lagged_corrs_f[l] > uci_f[l] or lagged_corrs_f[l] < lci_f[l]:
            ix_f.append(l)
```

```python
lags_array      = np.arange(-n_lags,n_lags+1)         # Array of lags for plotting
```

```python
fig     = plt.figure(figsize=(12, 6))
#lags_uf = plt.plot(lags_array/12,lagged_corrs_uf, '#fe6100', label = 'Unfiltered')
lags_f  = plt.plot(lags_array/12,lagged_corrs_f, '#fe6100', label = 'Two Layer Model')

zline   = plt.axhline(0, color='k', linewidth=2)
yline   = plt.axvline(0, color='k', linewidth=1, linestyle = '--')

#lci_  = plt.plot(lags_array/12,lci_uf, color='#fe6100',alpha=0.5)
#uci_  = plt.plot(lags_array/12,uci_uf, color='#fe6100',alpha=0.5)
#ci_sh = plt.fill_between(lags_array/12,lci_uf,uci_uf,color='#fe6100',alpha=0.2)

lci_  = plt.plot(lags_array/12,lci_f, color='#fe6100',alpha=0.5)
uci_  = plt.plot(lags_array/12,uci_f, color='#fe6100',alpha=0.5)
ci_sh = plt.fill_between(lags_array/12,lci_f,uci_f,color='#fe6100',alpha=0.2)

#plt.scatter(lags_array[ix_uf]/12, lagged_corrs_uf[ix_uf], s=25, color = '#fe6100', edgecolors='k',zorder = 10)
plt.scatter(lags_array[ix_f]/12, lagged_corrs_f[ix_f], s=25, color = '#fe6100',edgecolors='k',zorder = 10)

xtix = plt.yticks(fontsize=18)
ytix = plt.xticks(np.arange(-10,11,2),fontsize=18)
ylab = plt.ylabel('Correlation Coefficient', fontsize = 18)
xlab = plt.xlabel('Lag [Year]',fontsize = 18)
plt.xlim(lags_array[0]/12,lags_array[-1]/12)
plt.ylim(-0.7,0.7)
plt.legend(fontsize = 18)
plt.savefig(fig_path + 'lead_lag_nao_gs_two_layer.png', format='png', bbox_inches="tight",dpi=500)
```

```python
def moving_corr(nao_f, gsi_f, lag_gs = 7, chunk_sz = 48):
    corrs_ = []
    uci_ = []; lci_ = [];
    for c in range(len(gsi_f)-(chunk_sz+lag_gs)):
        corrs_.append(np.corrcoef(nao_f[c:c+chunk_sz],gsi_f[lag_gs+c:lag_gs+c+chunk_sz])[0,1])
        lci_temp, uci_temp = get_cis(nao_f[c:c+chunk_sz],gsi_f[lag_gs+c:lag_gs+c+chunk_sz],0)
        lci_.append(lci_temp)
        uci_.append(uci_temp)
    return(corrs_, lci_, uci_)
```

```python
chunk_sz = 48
lag_gs   = 7
corrs_nao, lci_nao, uci_nao = moving_corr(nao_f, gsi_f, lag_gs, chunk_sz)
```

```python
lci_nao = np.squeeze(np.stack(lci_nao)); uci_nao = np.squeeze(np.stack(uci_nao))
```

```python
time_plot = np.arange(np.datetime64(str(1993) + '-01-01'), np.datetime64(str(2023 + 1) + '-01-01'),
                 np.timedelta64(1, 'M'), dtype='datetime64[M]')
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
ix_nao = []
for l in range(len(corrs_nao)):
    if corrs_nao[l] > uci_nao[l] or corrs_nao[l] < lci_nao[l]:
        ix_nao.append(l)
ix_nao = np.array(ix_nao)
corrs_nao = np.array(corrs_nao)

av_per    = len(gsi_f)-len(corrs_nao)-lag_gs
plt.axhline(0, color='k', linewidth=1.5)
plt.plot(gsi_f.time[av_per//2:-av_per//2-lag_gs],corrs_nao,color = '#fe6100', label = 'NAO-GS')
plt.fill_between(gsi_f.time[av_per//2:-av_per//2-lag_gs],lci_nao,uci_nao,color = '#fe6100',alpha = 0.2)
plt.scatter(gsi_f.time[av_per//2:-av_per//2-lag_gs][ix_nao], corrs_nao[ix_nao], s=20, color = '#fe6100',edgecolors='k',zorder = 10)

#plt.plot(gsi_f.time[av_per//2:-av_per//2-lag_gs],corrs_eap,color = '#fe6100', label = 'EAP-GS')
#plt.fill_between(gsi_f.time[av_per//2:-av_per//2-lag_gs],lci_eap,uci_eap,color = '#fe6100',alpha = 0.2)
#plt.scatter(gsi_f.time[av_per//2:-av_per//2-lag_gs][ix_eap], corrs_eap[ix_eap], s=20, color = '#fe6100',edgecolors='k',zorder = 10)

plt.xticks(time_plot[2*12::5*12], labels = np.array(('1995','2000','2005','2010','2015','2020')), fontsize = 14)
plt.yticks(np.arange(-1,1.25,0.25), labels = np.array(('-1.00','-0.75','-0.50','-0.25','0.00','0.25','0.50','0.75','1.00')), fontsize = 14)
plt.xlim(gsi_f.time[0],np.datetime64(str(2023) + '-01-01'))
plt.ylim(-1,1)
plt.legend(fontsize=16)
#plt.title('EAP & NAO Moving Correlations w GSI (r = ' + str('{:.2f}'.format(np.corrcoef(corrs_nao,corrs_eap)[0,1])) + ')', fontsize = 16, fontweight='bold')
plt.savefig(fig_path + 'nao_lag_moving_average_two_layer.png', format='png', bbox_inches="tight",dpi=500)
```
